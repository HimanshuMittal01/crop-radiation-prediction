```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cowpea = pd.read_excel('../data/prepared.xlsx', sheet_name='Cowpea')
maize = pd.read_excel('../data/prepared.xlsx', sheet_name='Maize')
rice = pd.read_excel('../data/prepared.xlsx', sheet_name='Rice')
chickpea = pd.read_excel('../data/prepared.xlsx', sheet_name='Chickpea')
mustard = pd.read_excel('../data/prepared.xlsx', sheet_name='Mustard')
```

```python
sns.lmplot(x='GSR', y='Rn', data=cowpea)
sns.lmplot(x='GSR', y='Rn', data=maize)
sns.lmplot(x='GSR', y='Rn', data=rice)
sns.lmplot(x='GSR', y='Rn', data=chickpea)
sns.lmplot(x='GSR', y='Rn', data=mustard)
```

```python
sns.lmplot(x='CT', y='Rn', data=cowpea)
sns.lmplot(x='CT', y='Rn', data=maize)
sns.lmplot(x='CT', y='Rn', data=rice)
sns.lmplot(x='CT', y='Rn', data=chickpea)
sns.lmplot(x='CT', y='Rn', data=mustard)
```

```python
data = pd.concat([cowpea, rice, maize, chickpea, mustard], axis=0).reset_index(drop=True)
```

```python
data.loc[:, 'Time'] = data.loc[:, 'Time'].apply(lambda x: x.hour)
```

```python
data.loc[:, 'timesin'] = np.sin(data.loc[:, 'Time'] * (2 * np.pi) / 24)
data.loc[:, 'timecos'] = np.cos(data.loc[:, 'Time'] * (2 * np.pi) / 24)
```

```python
sns.heatmap(data.corr(), annot=True)
```

```python
for t in data['timecos'].value_counts().keys():
    sns.lmplot(x='GSR', y='Rn', data=data[data['timecos']==t])
```
